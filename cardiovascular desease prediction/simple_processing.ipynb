Exploratory data analysis


```python
import matplotlib.pyplot as plt
import seaborn as sns

#analiza klasnih tipova
fig,ax=plt.subplots(3,3,figsize=(20,10))

sns.countplot(x="Exercise",data=df,ax=ax[0,0])
sns.countplot(x="Heart_Disease",data=df,ax=ax[0,1])
sns.countplot(x="Skin_Cancer",data=df,ax=ax[0,2])

sns.countplot(x="Other_Cancer",data=df,ax=ax[0,2])
sns.countplot(x="Depression",data=df,ax=ax[1,0])


sns.countplot(x="Diabetes",data=df,ax=ax[1,1]) 
ax[1,1].tick_params(axis='x',rotation=90)

sns.countplot(x="Arthritis",data=df,ax=ax[1,2])

sns.countplot(x="Sex",data=df,ax=ax[2,0])
sns.countplot(x="Smoking_History",data=df,ax=ax[2,1])

sns.countplot(x="Age_Category",data=df,ax=ax[2,2])
ax[2,2].tick_params(axis='x',rotation=90)
```
